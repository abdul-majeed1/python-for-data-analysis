## **Vectorization**

```python
# Vectorization = perform operations on whole arrays at once without using loops.
```

```python
# Pros:
# 1. Faster
#       Operates on whole arrays at once → no explicit Python loops
# 2. Cleaner code
#       Less code, easier to read
# 3. Leverages NumPy optimizations
#       NumPy uses compiled C code under the hood
```

```python
# Cons:
# 1. Memory usage
#         Operates on whole arrays → can use more memory for large arrays
# 2. Not always faster for very small arrays
#         Overhead of vectorization may outweigh the benefit
# 3. Limited flexibility
#         Complex logic might need loops or other approaches
```

```python
# Example 1: Without vectorization (using loop)
a = np.array([1,2,3,4,5])
b = np.zeros(5)
```

```python
for i in range(len(a)):
  b[i] = a[i] * 2
```

```python
print(b)
```

```python
# Example 2: With vectorization (using NumPy)
```

```python
a = np.array([1,2,3,4,5])
b = a * 2   # vectorized operation
```

```python
print(b)
```

```python
a = np.array([1,2,3,4,5])
```

```python
def square(x):
  return x * x
vfunc = np.vectorize(square)
```

```python
print(vfunc(a))
```
